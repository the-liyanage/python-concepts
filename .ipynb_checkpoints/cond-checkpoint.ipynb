```python
#to check a value is greater, less or equal to each other

x = int(input("what's the value of x? "))
y = int(input("what's the value of y? "))

if x < y:
    print("x is less than y")
if x > y:
    print("x is greater than y")
if x == y:
    print("x is equal to y")
```

```python
#To Check a number whether it is even or odd

x = int(input("What's the value of x?: "))

if x % 2 == 0:
    print("This is an Even number")
else:
    print("This is an Odd number")
```

```python
#Converting the upper statement to a function

def main():
    x = int(input("What's the value of x? "))
    odd_even(x)

def odd_even(num):
    if num % 2 == 0:
        print("This is an Even number")
    else:
        print("This is an Odd number")

main()
```

```python
#Optimizing 1.0 

def main():
    x = int(input("What's the value of x? "))
    if odd_even(x):
        print("Even")
    else:
        print("Odd")
        
def odd_even(num):
    return True if num % 2 == 0 else False

main()
```

```python
#Optimizing 2.0 

#Since we are using Boolean values as the output

def main():
    x = int(input("What's the value of x? "))
    if odd_even(x):
        print("Even")
    else:
        print("Odd")

def odd_even(num):
    return num % 2 ==0

main()
```

```python
name = input("Enter your name:").lower()


if name == "lewis":
    print("Team Ferrari")
elif name == "charles":
    print("Team Ferrari")
elif name == "kimi":
    print("Team Mercedes")
else:
    print("Urm who?")
```

```python
#Optimizing 1.0 

name = input("Enter your name:").lower()

match name:
    case "lewis" | "charles":
        print("Team Ferrari")
    case "kimi":
        print("Team Mercedes")
    case _:
        print("Urm WHO?")
        
```
