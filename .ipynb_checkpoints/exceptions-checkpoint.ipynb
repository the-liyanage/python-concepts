```python
try:
    x = int(input("What's the value of x? "))
    print(f"x is {x}")
except ValueError:
    print("x is not an integer")            
```

```python
try:
    x = int(input("What's the value of x? "))
except ValueError:
    print("x is not an integer")
print(f"x is {x}")
```

```python
#To eliminate the above NameError we can use 'else'

try:
    x = int(input("What's the value of x? "))
except ValueError:
    print("x is not an integer")

else:
    print(f"x is {x}")
```
