```python
try:
    x = int(input("What's the value of x? "))
    print(f"x is {x}")
except ValueError:
    print("x is not an integer")            
```

```python
try:
    x = int(input("What's the value of x? "))
except ValueError:
    print("x is not an integer")
print(f"x is {x}")
```

```python
#To eliminate the above NameError we can use 'else'

try:
    x = int(input("What's the value of x? "))
except ValueError:
    print("x is not an integer")

else:
    print(f"x is {x}")
```

```python
while True:
    try:
        x = int(input("What's the value of x? "))
    except ValueError:
        print("x is not an integer")
    else:
        break
        
print(f"x is {x}")
```

```python
#Converting this to a function

def main():
    x = get_int()
    print(f"x is {x}")

def get_int():
    while True:
        try:
            x = int(input("What's the value of x? "))
        except ValueError:
            print("x is not an integer")
        else:
            break
    return x

main()
```

```python
#Optimizing the function 1.0

def main():
    x = get_int()
    print(f"x is {x}")

def get_int():
    while True:
        try:
            x = int(input("What's the value of x? "))
        except ValueError:
            print("x is not an integer")
        else:
            return x

main()
```

```python
#Optimizing the function 2.0

def main():
    x = get_int()
    print(f"x is {x}")

def get_int():
    while True:
        try:
            x = int(input("What's the value of x? "))
            return x
        except ValueError:
            print("x is not an integer")


main()
```

```python
#Optimizing the function further 3.0 

def main():
    x = get_int()
    print(f"x is {x}")

def get_int():
    while True:
        try:
            return int(input("What's the value of x? "))
        except ValueError:
            print("x is not an integer")

main()
```

```python
#Optimizing the function ----> using 'pass'

def main():
    x = get_int()
    print(f"x is {x}")

def get_int():
    while True:
        try:
            return int(input("What's the value of x? "))
        except ValueError:
            pass
main()
```

```python
#Optimizing the function further 4.0

def main():
    x = get_int("What's the value of x?")
    print(f"x is {x}")

def get_int(prompt):
    while True:
        try:
            return int(input(prompt))
        except ValueError:
            pass

main()
```
