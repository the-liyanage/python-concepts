```python
#Using the round function 

x = float(input("What's the value of x? "))
y = float(input("What's the value of y? "))

z = round(x+y)
print(f"{z}")
```

```python
#Seperate the value with (,)
x = float(input("what's the value of x? "))
y = float(input("what's the value of y? "))

z = round(x+y)
print(f"{z:,}")
```

```python
#Using the round function , to a specific number of decimals 

x = float(input("what's the value of x? "))
y = float(input("what's the value of y? "))

z = x/y

print(f"{z:.2f}")
```

```python
def main():
    num = get_number()
    say_forza(num)
    
def get_number():
    while True:
        num = int(input("What's the value of n? "))
        if num > 0:
            return num

def say_forza(num):
    for i in range(num):
        print("FORZA FERRARI")

main()
```

```python
def main():
    height, width = get_two()
    print_bricks(height,width)

def get_two():
    print("What are the values for the height and width, you propose? ")
    while True:
       height = int(input("height: "))
       width = int(input("width: "))
       if height > 0 and width > 0:
            return height, width

def print_bricks(height,width):
    #this represents printing each row 
    for i in range(height):

        #this represents printing each brick in the row
        for i in range(width):

            #print brick
            print("#", end=" ")

        #after printing n number of width go to the next row
        print("\n")    
        
          
main()
```
