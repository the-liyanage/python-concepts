```python
#Over-writing the print function

name = input("What's your name?")
print("hi,", end="")
print(name)
```

```python
#Over-writing the print function(2.0)

name = input("What's your name?")
#positional parameters(name)          named parameters(sep,end)
print("hi,", name, sep="??")
```

```python
#Using the round function 

x = float(input("What's the value of x? "))
y = float(input("What's the value of y? "))

z = round(x+y)
print(f"{z}")
```

```python
#Seperate the value with (,)
x = float(input("what's the value of x? "))
y = float(input("what's the value of y? "))

z = round(x+y)
print(f"{z:,}")
```

```python
#Using the round function , to a specific number of decimals 

x = float(input("what's the value of x? "))
y = float(input("what's the value of y? "))

z = x/y

print(f"{z:.2f}")
```
