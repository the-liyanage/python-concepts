```python
import random
```

```python
coin = random.choice(["heads", "tails"])
print(coin)
```

```python
#Choosing a random integer
a = random.randint(1,10)
print(a)
```

```python
#Shuffling a list (this function takes a list as an argument)
b =['hey', 'hi', 'no']
random.shuffle(b)

for word in b:
    print(word)
```
