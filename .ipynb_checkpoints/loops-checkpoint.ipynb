```python
#To get a specific value (postiive n value)

while True:
    n = int(input("what's the value of n? "))
    if n < 0:
        continue
    else:
        break
```

```python
while True:
    n = int(input("What's the value of n? "))
    if n > 0:
        break
for _ in range(n):
    print("hi")
```

```python
#Optimizing this with a fucntion

def main():
    while True:
        n = int(input("What's the value of n? "))
        if n > 0:
            break
    for _ in range(n):
        print("hi")


main()
    
```

```python
def main():
    while True:
        n = int(input("What's the value of n? "))
        if n > 0:
            break 

    for i in range(n):
        return "HI"

main()

#print() → shows something on the screen
#return → exits the function and sends a value back
#return inside a loop stops the loop immediately
```
