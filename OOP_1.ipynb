#### 5. Introducing Methods(Behaviours)

```python
class Student():
    def __init__(self,name,house,charm):
        if not name:
            raise ValueError("Missing name")
        if house not in ["Gryffindor", "Hufflepuff", "Ravenclaw", "Slytherin"]:
            raise ValueError("Invalid Error")
        self.name = name
        self.house = house

    def __str__(self):
        return f"{self.name} from {self.house}"

    def charm(self):
        #Matching the charms with the emojis
        match self.charm:
            case "Stag":
                return "ðŸ¦„"
            case "Otter":
                return "ðŸ¦¦"
            case "Jack":
                return "ðŸ™"
            #Giving a default value
            case _:
                return "ðŸ„"


def main():
    student = get_student()
    print("Expecto Patronum!")
    print(student.charm())

def get_student():
    name = input("Name: ")
    house = input("House: ")
    charm = input("Patronus: ")
    return Student(name,house,charm)

if __name__ == "__main__":
    main()
```

#### 5. Introducting the Getter and Setter 

- To prevent overwriting the attributes(values) -----> This will generate a ValueError

```python
class Student():
    def __init__(self,name,house):
        if not name:
            raise ValueError("Missing Values")
        self.name = name
        self.house = house

    def __str__(self):
        return f"{self.name} from {self.house}"

    #Getter
    #Function for a class that gets some attribute

    @property 
    def house(self):
        return self._house

    #Setter
    #Fucntion in some class that sets some value
    
    @house.setter
    def house(self,house):
        if house not in ["Gryffindor", "Hufflepuff", "Ravenclaw", "Slytherin"]:
            raise ValueError("Invalid House")
        self._house = house


def main():
    student = get_student()
    student.house = "New Drive"
    print(student)

def get_student():
    name = input("Name: ")
    house = input("House: ")
    return Student(name,house)

if __name__ == "__main__":
    main()
```

```python
import random

class Hat():
    def __init__(self):
        self.houses = ["Gryffindor", "Hufflepuff", "Ravenclaw", "Slytherin"]

    def sort(self,name):
        print(name, "is in", random.choice(self.houses))
        

hat = Hat()
hat.sort("Harry")
```
